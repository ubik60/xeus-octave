Följande två AR processer används som exempel i boken av Percival och Walden.

```octave
ar2 = [ 1 -0.75 0.5 ]; % AR(2) ekvation  (45) s 45
ar4 = [ 1 -2.7607 3.8106 -2.6535 0.9238 ]; % AR(4) ekvation (46a) 
```

Här plottas motsvarande överföringsfunktioner

```octave
K = 128;
[Sar4 fs] = freqz([1],ar4,128,1);
[Sar2 fs] = freqz([1],ar2,128,1);
```

```octave
freqz_plot(fs, Sar4)
```

```octave
freqz_plot(fs, Sar2)
```

```octave
N = 2*K+1;
xar4 = filter([1],ar4,randn(1,N));
xar2 = filter([1],ar2,randn(1,N));
```

```octave
Sp4 = fft(xar4)(1:length(fs))/sqrt(N);
Sp2 = fft(xar2)(1:length(fs))/sqrt(N);
```

```octave
plot(fs,pow2db(abs(Sar4).^2),'b-',fs,pow2db(abs(Sp4).^2),'r+-')
legend("Exakt AR(4)", "Periodogram");
```

```octave
plot(fs,abs(Sar4).^2,'b-',fs,abs(Sp4).^2,'r+-')
```
